```python
def recommended_course(age,stream,interest):
  if stream=='Science':
    if interest=='High':
      return np.random.choice(['AI','Data Science'])
    elif interest=='Medium':
      return np.random.choice(['Python','AI'])
    else:
      return 'Python'
  elif stream=='Commerce':
    if interest=='High':
      return 'Economics'
    elif interest=='Medium':
      return 'Accounting'
    else:
      return 'Management'
  elif stream=='Arts':
    if interest=='High':
      return 'Sociology'
    elif interest=='medium':
      return 'History'
    else:
      return 'History'
  else:
    np.random.choice(courses)

```

```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,classification_report
```

```python
ages = [18, 19, 20, 21, 22, np.nan]
streams = ['Science', 'Commerce', 'Arts', None]
interest_levels = ['High', 'Medium', 'Low']
courses = ['Python', 'Accounting', 'Data Science', 'Sociology', 'Economics', 'History', 'AI', 'Management']
```

```python
data = {
    'Age': [np.random.choice(ages) for _ in range(1000)],
    'Stream': [np.random.choice(streams) for _ in range(1000)],
    'Interest_Level': [np.random.choice(interest_levels) for _ in range(1000)]
}

```

```python
df = pd.DataFrame(data)
df['Stream'] = df['Stream'].replace({None: np.nan})
```

```python
df.isnull().sum()
```

```python
imputer = SimpleImputer(strategy='most_frequent')
df[['Age','Stream']] = imputer.fit_transform(df[['Age','Stream']])
```

```python
df['Recommend_Course']=df.apply(lambda x:recommended_course(x['Age'],x['Stream'],x['Interest_Level']),axis=1)
df.head(5)
```

```python
df.isnull().sum()
```

```python
le_stream=LabelEncoder()
le_interest=LabelEncoder()
le_course=LabelEncoder()
```

```python
df['Interest_Level'].head()
```

```python
le_interest.fit_transform(df['Interest_Level'])
```

```python
df['Stream_label']=le_stream.fit_transform(df['Stream'])
df['Interest_Level_label']=le_interest.fit_transform(df['Interest_Level'])
df['Recommend_Course_label']=le_course.fit_transform(df['Recommend_Course'])

```

```python
df['Stream'].unique()
```

```python
# unique_values = [str(val) for val in df['Interest_Level'].unique()]
# print(unique_values)

```

```python
df.head()
```

```python
X=df[['Age','Stream_label','Interest_Level_label']]
y=df['Recommend_Course']
```

```python
df['Stream'].unique()
```

```python
scaled=StandardScaler()
X_scaled=scaled.fit_transform(X)
X_scaled
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,train_size=0.8,random_state=42)
```

```python
param_grid = {'n_neighbors': np.arange(1, 11)}
knn = KNeighborsClassifier()
grid = GridSearchCV(knn, param_grid,cv=3)
grid.fit(X_train, y_train)

best_k = grid.best_params_['n_neighbors']
print("🔍 Best K:", best_k)
```

```python
knn_best=KNeighborsClassifier(n_neighbors=best_k)
knn_best.fit(X_train,y_train)
```

```python
y_pred=knn_best.predict(X_test)
y_pred
```

```python
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
print("✅ Accuracy:", accuracy_score(y_test, y_pred))

# print("📊 Classification Report:\n", classification_report(y_test, y_pred, labels=unique_labels, target_names=target_names))

# print("📊 Classification Report:\n", classification_report(y_test, y_pred, target_names=le_course.classes_))
```
